```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('coiffeurs.xlsx')
df = df.dropna()
df
```

```python
#CorrÃ©lation Nord-Sud :
#DonnÃ©es : Plus haute latitude France : 51.05 ; Plus basse latitude France : 41.22 
categorieLatitude = pd.cut(df['lat'], bins=[41.22 + k*0.25 for k in range (40)], labels = [str(i) for i in range(39)])
df['CateLat'] = categorieLatitude
L = df.groupby(['CateLat']).size()
L
```

```python
fig = plt.figure()
X = [ 41.22 + k*0.25 for k in range (39)]
x = X
height = L
width = 1.0

plt.bar(x, height, width, color='b')
plt.title('Nombre de coiffeurs drÃ´les en fonction de la latitude')
plt.show()
```

```python
categorieLongitude = pd.cut(df['lng'], bins=[4.47 + k*0.1 for k in range (37)], labels = [str(i) for i in range(36)])
df['CateLng'] = categorieLongitude
M = df.groupby(['CateLng']).size()
M
```

```python
fig2 = plt.figure()
X = [4.47 + k*0.1 for k in range (36)]
x = X
height = M
width = 1.0

plt.bar(x, height, width, color='b')
plt.title('Nombre de coiffeurs drÃ´les en fonction de la longitude')
plt.show()
```
