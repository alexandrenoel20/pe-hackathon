```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('coiffeurs.xlsx')
df
```

```python
df.info()
```

```python
df = df.dropna(axis=0, how='any')
df
```

```python
df = df.drop(['type','markerinnerhtml','liinnerhtml','type.1'],axis=1)
df.head(3)
```

```python
df['nom']=df['nom'].astype(str)
df.dtypes
```

```python
df['hair'] = df['nom'].str.contains('hair')
df
```

```python
df['tif'] = df['nom'].str.contains('tif')
df
```

```python
df['coif'] = df['nom'].str.contains('coif')
df
```

```python
df['nb-jeu-de-mots']= df['tif']*1+df['hair']*1+df['coif']*1
df.value_counts('nb-jeu-de-mots')
```
