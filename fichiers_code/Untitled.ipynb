```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('coiffeurs.xlsx')
df = df.dropna()
df.drop(df.loc[df['num'] == '[ND]' ].index, inplace=True)
df
```

```python
categorieNum = pd.cut(df['num'], bins=[10*k for k in range (51)])
df['CateNum'] = categorieNum
L = df.groupby(['CateNum']).size()
```

```python
fig = plt.figure()
X = [k for k in range (501)]
x = X
height = L
width = 1.0

plt.bar(x, height, width, color='b')
plt.title('Nombre de coiffeurs drôles en fonction du mu')
plt.show()
```
